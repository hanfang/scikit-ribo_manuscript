### 1. TE comparsion: skr vs. rpkm

#### (1) Low abundance genes' TEs have high dispersion 

```python
# skr
skr_weinberg = pd.read_table(skr_weinberg_fn, header=0)
plt.clf()
ax1 = sns.distplot(skr_weinberg['log2_TE'])
ax1.set(xlim=(-8, 8))
ax1.axvline(0, color="#999999", dashes=[3,2], zorder=-1)
plt.gcf()
plt.savefig("../Figures/skr_weinburg_log2TE.pdf")
#skr_weinberg.head
```

```python
# read data
kallisto_riboseq = pd.read_table(kallisto_riboseq_fn, header=0)
kallisto_rnaseq = pd.read_table(kallisto_rnaseq_fn, header=0)

# compute TE from rpkm
kallisto_riboseq["rpkm_ribo"] = kallisto_riboseq["est_counts"] * (10 ** 9) / (kallisto_riboseq["length"] * np.sum(kallisto_riboseq["est_counts"]))
kallisto_rnaseq["rpkm_rna"] = kallisto_rnaseq["est_counts"] * (10 ** 9) / (kallisto_rnaseq["length"] * np.sum(kallisto_rnaseq["est_counts"]))
rkpm_weinburg = pd.merge(kallisto_riboseq, kallisto_rnaseq, on="target_id", suffixes=["_ribo","_rna"])
rkpm_weinburg = rkpm_weinburg [(rkpm_weinburg["tpm_ribo"] > 0) & (rkpm_weinburg['tpm_rna'] > 0)]
rkpm_weinburg['log2_TE'] = np.log2(rkpm_weinburg["rpkm_ribo"] / rkpm_weinburg["rpkm_rna"])
rkpm_weinburg["log2_TE"] = (rkpm_weinburg["log2_TE"] - np.median(rkpm_weinburg["log2_TE"])) / np.log(2)

# plot
plt.clf()
ax1 = sns.distplot(rkpm_weinburg['log2_TE'])
ax1.axvline(0, color="#999999", dashes=[3,2], zorder=-1)
#ax1.set(xlim=(-8, 8))
plt.gcf()
plt.savefig("../Figures/rkpm_weinburg_log2TE.pdf")
```

```python
rkpm_weinburg["abundance"] = np.where( (rkpm_weinburg["tpm_ribo"] > 10) & (rkpm_weinburg["tpm_rna"] > 10), "Yes", "No") 

# violin plot - log2 TE
plt.clf()
vio0 = sns.violinplot(x="abundance", y="log2_TE", data=rkpm_weinburg)
vio0.set(ylim= [-60,60])
vio0.set_xticklabels(['Low','High'])
vio0.set_ylabel('log2(TE) from RPKM')
plt.tight_layout()
plt.gcf()
plt.savefig("../Figures/weinburg_log2TE_rkpm_abundance.pdf")
```

```python
# combine data
TE_combine = pd.merge(skr_weinberg, rkpm_weinburg, left_on=["gene"], right_on=["target_id"], suffixes=["_skr","_rpkm"])
TE_excluded = pd.merge(rkpm_weinburg, skr_weinberg, left_on=["target_id"], right_on=["gene"], suffixes=["_rpkm","_skr"], how="left")
TE_excluded["class"] = np.where(TE_excluded["gene"].isnull(), "TPM < 1", "TPM > 1") 

# violin plot - log2 TE
plt.clf()
vio1 = sns.violinplot(x="class", y="log2_TE_rpkm", data= TE_excluded)
vio1.set(ylim= [-60,60])
vio1.set_xticklabels(['TPM < 1\n','TPM > 1'])
vio1.set_ylabel('log2(TE) based on RPKM')
vio1.set_xlabel(' ')
plt.tight_layout()
plt.gcf()
plt.savefig("../Figures/weinburg_log2TE_rkpm_unique.pdf")
```

#### (2) Pairwise comparisons showed that scikit-ribo:
#####     a) is robust to low abudance genes 
#####     b) corrected biases due to codons and 2' structure

```python
TE_combine["log2_TE_rpkm_adj"] = np.where(TE_combine["log2_TE_rpkm"] < -10, -9.5, TE_combine["log2_TE_rpkm"]) 
TE_combine['deltaTE'] = TE_combine["log2_TE_skr"] - TE_combine["log2_TE_rpkm"]

#TE_combine[(TE_combine["log2_TE_rpkm"]>10) | (TE_combine["log2_TE_rpkm"]<-10)]
```

```python
plt.clf()
TE_corr = round(pearsonr(TE_combine["log2_TE_rpkm"], TE_combine["log2_TE_skr"])[0] ,2)

g1 = sns.jointplot(x="log2_TE_rpkm_adj", y="log2_TE_skr", data=TE_combine, 
                    xlim= [-10,10], ylim= [-10,10], color='white', marker='.', 
                   s=40, alpha=0, size=6, stat_func=None)
g1 = (g1.set_axis_labels("RPKM log2(TE) (median normalized)", "scikit-ribo log2(TE)"))
x0, x1 = g1.ax_joint.get_xlim()
y0, y1 = g1.ax_joint.get_ylim()
lims = [max(x0, y0), min(x1, y1)]
g1.ax_joint.plot(lims, lims, ':k')

upper = TE_combine[TE_combine['deltaTE'] > 0.5]
g1.x, g1.y = upper.log2_TE_rpkm, upper.log2_TE_skr
g1.plot_joint(plt.scatter, marker='.', c=snsGreen, s=30, alpha=0.5)

lower = TE_combine[TE_combine['deltaTE'] < -0.5]
g1.x, g1.y = lower.log2_TE_rpkm, lower.log2_TE_skr
g1.plot_joint(plt.scatter, marker='.', c=snsRed , s=40, alpha=0.5)


middle = TE_combine[(TE_combine['deltaTE'] >= -0.5) & (TE_combine['deltaTE'] <= 0.5)]
g1.x, g1.y = middle.log2_TE_rpkm, middle.log2_TE_skr
g1.plot_joint(plt.scatter, marker='.', c=snsBlue, s=40, alpha=0.5)

outlier = TE_combine[TE_combine["log2_TE_rpkm"] < -10]
g1.x, g1.y = outlier.log2_TE_rpkm_adj, outlier.log2_TE_skr
g1.plot_joint(plt.scatter, marker='x', c="black", s=40, alpha=0.5)

plt.text(-8, 8, r"$r$ = " + str(TE_corr), fontsize=20)

plt.gcf()
plt.savefig("../Figures/compare_skr_rpkm_TE.pdf")
```
